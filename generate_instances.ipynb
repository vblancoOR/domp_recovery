```python
#from domp_lib import *
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import math

#Numero de puntos
N=[10,20,30,40,50,100,200,500]
#Numero de bolas
M=[1,2,3]
# Dimension
D = [2,3]
#Separacion entre centros
Sep=[2,2.5,3,3.5,4,4.5,5]
```

```python
list_files=[]
for n in N:
    for m in M:
        for d in D:
            for sep in range(len(Sep)):
                for it in range(5):
                    sample_size05 = max(1, int(0.05 * n))
                    sample_size10 = max(1, int(0.10 * n))
                    sample_size15 = max(1, int(0.15 * n))
                    sample05 = np.random.choice(range(n), size=sample_size05, replace=False)
                    sample10 = np.random.choice(range(n), size=sample_size10, replace=False)
                    sample15 = np.random.choice(range(n), size=sample_size15, replace=False)
                    nb = n//m
                    y=[]
                    theta1 = np.random.uniform(0, 2*math.pi, size=n)
                    r = np.random.uniform(0, 1, size=n)
                    r5=r.copy()
                    r10=r.copy()
                    r15=r.copy()
                    for i in range(n):
                        if i in sample05:
                            r5[i]=1+np.random.uniform(low=0,high=Sep[sep]-2+0.2)
                        if i in sample10:
                            r10[i]=1+np.random.uniform(low=0,high=Sep[sep]-2+0.2)
                        if i in sample15:
                            r15[i]=1+np.random.uniform(low=0,high=Sep[sep]-2+0.2)
                        for j in range(nb):
                            if i%nb==j:
                                y.append(j)
                    
                    if d==2:
                        #print(n, m, Sep[sep], d)
                        X = np.array([[Sep[sep]*y[i] + r[i]*np.cos(theta1[i]), r[i]*np.sin(theta1[i])] for i in range(n)])
                        X05 = np.array([[Sep[sep]*y[i] + r5[i]*np.cos(theta1[i]), r5[i]*np.sin(theta1[i])] for i in range(n)])
                        X10 = np.array([[Sep[sep]*y[i] + r10[i]*np.cos(theta1[i]), r10[i]*np.sin(theta1[i])] for i in range(n)])
                        X15 = np.array([[Sep[sep]*y[i] + r15[i]*np.cos(theta1[i]), r15[i]*np.sin(theta1[i])] for i in range(n)])
                        # fig, ax = plt.subplots()
                        # # Define a circle at (x, y) with radius r
                        # circle = patches.Circle((0, 0), 1, edgecolor='blue', facecolor='none', linewidth=2)
                        # ax.add_patch(circle)
                        # circle = patches.Circle((Sep[sep], 0), 1, edgecolor='blue', facecolor='none', linewidth=2)
                        # ax.add_patch(circle)
                        # circle = patches.Circle((2*Sep[sep], 0), 1, edgecolor='blue', facecolor='none', linewidth=2)
                        # ax.add_patch(circle)
                        # ax.scatter(X10[:,0], X10[:,1], c=y)
                        # # Set axis limits and aspect ratio
                        # ax.set_xlim(-1.5, 2*Sep[sep]+1.5)
                        # ax.set_ylim(-1.5, 1.5)
                        # ax.set_aspect('equal')
                        # ax.scatter(X10[sample10,0], X10[sample10,1], c='r')
                        # plt.show()
                    elif d==3:
                        theta2 = np.random.uniform(0, math.pi, size=n)
                        X = np.array([[Sep[sep]*y[i] + r[i]*np.cos(theta1[i])*np.sin(theta2[i]), r[i]*np.sin(theta1[i])*np.sin(theta2[i]), r[i]*np.cos(theta2[i])] for i in range(n)])
                        X05 = np.array([[Sep[sep]*y[i] + r5[i]*np.cos(theta1[i])*np.sin(theta2[i]), r5[i]*np.sin(theta1[i])*np.sin(theta2[i]), r5[i]*np.cos(theta2[i])] for i in range(n)])
                        X10 = np.array([[Sep[sep]*y[i] + r10[i]*np.cos(theta1[i])*np.sin(theta2[i]), r10[i]*np.sin(theta1[i])*np.sin(theta2[i]), r10[i]*np.cos(theta2[i])] for i in range(n)])
                        X15 = np.array([[Sep[sep]*y[i] + r15[i]*np.cos(theta1[i])*np.sin(theta2[i]), r15[i]*np.sin(theta1[i])*np.sin(theta2[i]), r15[i]*np.cos(theta2[i])] for i in range(n)])
                    # Save the data
                    file0 = f"instance_n{n}_m{m}_s{sep}_d{d}_o00_{it}.txt"
                    np.savetxt(file0, X)
                    file5 = f"instance_n{n}_m{m}_s{sep}_d{d}_o05_{it}.txt"
                    np.savetxt(file5, X05)
                    file10 = f"instance_n{n}_m{m}_s{sep}_d{d}_o10_{it}.txt"
                    np.savetxt(file10, X10)
                    file15 = f"instance_n{n}_m{m}_s{sep}_d{d}_o15_{it}.txt"
                    np.savetxt(file15, X15)
```
